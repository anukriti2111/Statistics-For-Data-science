```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Laliga.csv',header=1)
```

```python
data.head()
```

```python
# Step 1: Read the data and replace dashes with 0
data.replace('-', 0, inplace=True)
```

```python
# Step 2: Print teams that started playing between 1930-1980
teams_between_1930_1980 = data[data['Debut'].str[:4].astype(int).between(1930, 1980)]
print("Teams that started playing between 1930-1980:")
print(teams_between_1930_1980['Team'])
```

```python
# Step 3: Print top 5 teams in terms of points
data['Points'] = data['Points'].astype(int)
top_5_teams = data.nlargest(5, 'Points')
print("Top 5 teams in terms of points:")
print(top_5_teams[['Team', 'Points']])
```

```python
# Step 4: Write Goal_diff_count function and find team with max and min goal difference
def Goal_diff_count():   
     data['Goal_diff_count'] = data['GoalsFor'].astype(int)-data['GoalsAgainst'].astype(int)
     # if require to get the absolute value below line can be used
     # data['Goal_diff_count'] = data['Goal_diff_count'].abs()
     return data[['Team','Goal_diff_count']].sort_values(by = 'Goal_diff_count',ascending=False)
    
Goal_diff_count()
```

```python
print("The team which has maximum goal difference is :")
Goal_diff_count().head(1)
```

```python
print("The team which has minimum goal difference is :")
Goal_diff_count().tail(1)
```

```python
# Step 5: Create Winning Percent column and print top 5 teams with highest Winning Percentage
data['GamesWon'] = data['GamesWon'].astype(int)
data['GamesPlayed'] = data['GamesPlayed'].astype(int)
data['WinningPercent'] = (data['GamesWon'] / data['GamesPlayed']) * 100
data['WinningPercent'].fillna(0, inplace=True)
```

```python
top_5_winning_percent_teams = data.nlargest(5, 'WinningPercent')
print("Top 5 teams with highest Winning Percentage:")
print(top_5_winning_percent_teams[['Team', 'WinningPercent']])
```

```python
# Step 6: Group teams by Best Position and sum their points
grouped_by_best_position = data.groupby('BestPosition').agg({'Points': 'sum', 'Team': ', '.join})
print("Sum of points for each Best Position:")
print(grouped_by_best_position)
```
